```python
import pandas as pd

p13 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/41467_2021_25583_MOESM9_ESM.xlsx',sheet_name=1)
p14 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/41467_2021_25583_MOESM9_ESM.xlsx',sheet_name=2)
```

```python
p13_value = p13.values.tolist()
p14_value = p14.values.tolist()
p13_list = [x[0] for x in p13_value][1:]
p14_list = [x[0] for x in p14_value][1:]
```

```python
p13_list
```

```python
p13_gather = []
p14_gather = []

line_number = 0
for line in open('/content/drive/MyDrive/Colab Notebooks/MuTHER_Fat_450K_norm_AE_030913.txt', 'r'):
  content = line.split()
  if content[0] in p13_list:
    p13_gather.append(content)
  if content[0] in p14_list:
    p14_gather.append(content)
  line_number += 1
  print(line_number, 'lines have been processed!')
```

```python
# p13_gather[0]
# p14_gather

p13_dict = {}
p14_dict = {}

for line in p13_gather:
  p13_dict[line[0]] = line

for line in p14_gather:
  p14_dict[line[0]] = line
```

```python
p13_sorted = []
p14_sorted = []

for id in p13_list:
  p13_sorted.append(p13_dict[id])

for id in p14_list:
  p14_sorted.append(p14_dict[id])
```

```python
header13 = []
header14 = []

for i in range(len(p13_sorted)):
  header13.append(p13_sorted[i][0])

for i in range(len(p14_sorted)):
  header14.append(p14_sorted[i][0])
```

```python
data13 = []
data14 = []

for i in range(len(p13_sorted[0])):
  temp = []
  for j in range(len(p13_sorted)):
    temp.append(p13_sorted[j][i])
  data13.append(temp)

for i in range(len(p14_sorted[0])):
  temp = []
  for j in range(len(p14_sorted)):
    temp.append(p14_sorted[j][i])
  data14.append(temp)
```

```python
data13 = data13[1:]
data14 = data14[1:]
```

```python
import csv

with open('/content/drive/MyDrive/Colab Notebooks/EMAT1866Supple13.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header13)
    writer.writerows(data13)

with open('/content/drive/MyDrive/Colab Notebooks/EMAT1866Supple14.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header14)
    writer.writerows(data14)
```
